```python
def calculate_threshold(df, delta=30, rate=0.45):
    df = get_future_price(df, delta=delta)

    # Tính tỷ lệ thay đổi giá
    df.loc[:, ['Return']] = df[f'Price_t_plus_{delta}'] / df['Close'] - 1

    # Loại bỏ giá trị NaN (do shift)
    df = df.dropna()

    # Tách return thành tăng (dương) và giảm (âm)
    positive_returns = df['Return'][df['Return'] > 0]  # Tăng giá

    std_positive = positive_returns.std()

    print(f'Delta = {delta}')
    
    # Đề xuất threshold 
    threshold = std_positive * rate # 1.5 lần độ lệch chuẩn cho BUY
    
    print(f"Threshold gợi ý: {threshold:.4%}")
    
    # Tính giá trị tiền dựa trên giá trung bình trong df
    avg_price = df['Close'].mean()
    print(f"\nGiá trung bình trong dữ liệu: ${avg_price:.2f}")

    # Tính giá trị tiền
    threshold_in_money = avg_price * threshold

    print(f"Threshold ({threshold}) thành tiền: ${threshold_in_money:.2f}")
    
    return delta, threshold
    
```

```python
delta, threshold = calculate_threshold(train_df, 1, 0.4)
```

```python
# Simple labeling
def simple_labeling(df, delta, threshold):
    df.loc[:, [f'Price_t_plus_{delta}']] = df['Close'].shift(-delta)

    # Tính tỷ lệ thay đổi giá
    df.loc[:, ['Return']] = df[f'Price_t_plus_{delta}'] / df['Close'] - 1

    # Loại bỏ giá trị NaN (do shift)
    df = df.dropna()
    
    # Kiểm tra phân bố nhãn với threshold riêng
    df.loc[:, ['Label']] = np.where(df[f'Price_t_plus_{delta}'] > df['Close'] * (1 + threshold), 'BUY',
                        np.where(df[f'Price_t_plus_{delta}'] < df['Close'] * (1 - threshold), 'SELL', 'HOLD'))

    label_distribution = df['Label'].value_counts(normalize=True)
    print("\nPhân bố nhãn với threshold riêng:")
    print(label_distribution)
    
    return df
```

```python
labeled_dev_df = simple_labeling(dev_df, delta, threshold)
# labeled_dev_df
```

```python
labeled_test_df = simple_labeling(test_df, delta, threshold)
# labeled_test_df
```

```python
def counter_label(df, delta):
    cnt = 0
    for i in range(0, len(df) - delta, delta):
        label = df['Label'].iloc[i-30]
        for j in range(i, i+delta):
            if label == 'BUY':
                if df['Label'].iloc[j] == 'SELL':
                    cnt += 1
                    break
            elif label == 'SELL':
                if df['Label'].iloc[j] == 'BUY':
                    cnt += 1
                    break
    return cnt


print("Tong so label: ", len(labeled_train_df))
cnt = counter_label(labeled_train_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(labeled_train_df) * 100)


print("Tong so label: ", len(labeled_dev_df))
cnt = counter_label(labeled_dev_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(labeled_dev_df) * 100)


print("Tong so label: ", len(labeled_test_df))
cnt = counter_label(labeled_test_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(labeled_test_df) * 100)
```

Tại thời điểm T, xét trong khoảng (T, T + $\Delta$):
- Nếu T đang là BUY (giá tại T + $\Delta$ cao hơn T), trong khoảng này xuất hiện điểm có giá nhỏ hơn T rất nhiều (đáy $\sqrt{}$)
        thì nhãn tại T sẽ được chuyển thành SELL.
- Và ngược lại

```python
def dynamic_labeling(df, delta, threshold):
    df = df.copy()
    
    # Bước gán nhãn ban đầu
    df = simple_labeling(df, delta, threshold)
    
    # Tính sẵn min/max trong cửa sổ kế tiếp delta dòng => Cực trị
    # Dịch ngược DataFrame, tính rolling min/max, rồi dịch đúng thứ tự
    reversed_close = df['Close'].iloc[::-1]
    rolling_min = reversed_close.rolling(window=delta, min_periods=1).min().iloc[::-1].shift(1)
    rolling_max = reversed_close.rolling(window=delta, min_periods=1).max().iloc[::-1].shift(1)

    # Áp dụng min/max tương ứng từng hàng
    df['future_min'] = rolling_min
    df['future_max'] = rolling_max

    # Cập nhật nhãn theo giá cực trị
    df.loc[(df['Label'] == 'BUY') & (df['future_min'] < df['Close'] * (1 - threshold)), 'Label'] = 'SELL'
    df.loc[(df['Label'] == 'SELL') & (df['future_max'] > df['Close'] * (1 + threshold)), 'Label'] = 'BUY'

    label_distribution = df['Label'].value_counts(normalize=True)
    print("\nPhân bố nhãn sau cập nhật:")
    print(label_distribution)

    return df
```

```python
dynamic_labeled_train_df = dynamic_labeling(train_df, delta, threshold)
# dynamic_labeled_train_df
```

```python
dynamic_labeled_dev_df = dynamic_labeling(dev_df, delta, threshold)
# dynamic_labeled_dev_df
```

```python
dynamic_labeled_test_df = dynamic_labeling(test_df, delta, threshold)
# dynamic_labeled_test_df
```

```python

print("Tong so label: ", len(dynamic_labeled_train_df))
cnt = counter_label(dynamic_labeled_train_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(dynamic_labeled_train_df) * 100)


print("Tong so label: ", len(dynamic_labeled_dev_df))
cnt = counter_label(dynamic_labeled_dev_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(dynamic_labeled_dev_df) * 100)


print("Tong so label: ", len(dynamic_labeled_test_df))
cnt = counter_label(dynamic_labeled_test_df, delta)
print("So lan xuat hien nhan nguoc: ", cnt)
print("Ti le xuat hien nhan nguoc", cnt / len(dynamic_labeled_test_df) * 100)
```
